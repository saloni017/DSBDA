```python
import numpy as np
```

```python
import pandas as pd
```

```python
##Importing Dataset Iris 
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1 ## Reading dataframes
```

```python
data1.isnull().sum() ##Checking null values
```

```python
data1.dtypes ##Checking data types
```

```python
data1.info ##Checking information of data
```

```python
from sklearn.model_selection import train_test_split ##importing library from sklearn
```

```python
X=data1.drop("target",axis=1)
Y=data1["target"] ##Declaring X and Y separately
```

```python
X ##Printing X
```

```python
Y ##Printing Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=100)
##Splitting 
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler ##importing MinMaxScaler for Scaling
```

```python
ms = MinMaxScaler() ##Declaring MinMaxScaler
```

```python
x_train = ms.fit_transform(x_train) ##Scaling x_train
```

```python
x_train
```

```python
x_test = ms.fit_transform(x_test) ##Scaling x_test
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB ##importing Naive Bayes
```

```python
Nb_model = GaussianNB() ##declaring Gaussianmodel
```

```python
Nb_model.fit(x_train,y_train)
```

```python
y_pred = Nb_model.predict(x_test) ##Predicting Y
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix 
##importing these libraries for accuracy score
```

```python
ascore = accuracy_score(y_pred,y_test)
```

```python
ascore
```

```python
cm = confusion_matrix(y_pred,y_test)
```

```python
cm
```

```python
pscore= precision_score(y_pred,y_test,average='macro')
```

```python
pscore
```

```python
rscore = recall_score(y_pred,y_test,average='macro')
```

```python
rscore
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
discp = ConfusionMatrixDisplay(cm) ##Plotting graph
```

```python
discp.plot()
```
